```python
import pandas as pd
```

```python
data = pd.read_csv('input.txt', delim_whitespace=True , names=['mov','meters'])
data
```

```python
total_foward = data['meters'][data['mov'] == 'forward']
total_foward = total_foward.sum()
total_foward
```

```python
total_up = data['meters'][data['mov'] == 'up']
total_up = total_up.sum()
total_up
```

```python
total_down = data['meters'][data['mov'] == 'down']
total_down = total_down.sum()
total_down
```

```python
total_vertical = abs(total_up - total_down)
```

```python
result = total_vertical * total_foward
result
```

down X increases your aim by X units.  
up X decreases your aim by X units.  

forward X does two things:  
It increases your horizontal position by X units.  
It increases your depth by your aim multiplied by X.  

This can be resolve by thinking this as a 3d vector which coordinates are (h,d,a) h:   horizontal pos, d: depth, a: aim   

the following is resolved as:   

h = foward_X 

d = foward_X * aim

a = aim

in vector form: 
(foward_X,foward_X * aim, aim)

```python
data['aim'] = 0
```

```python
aim = 0 
d = 0
for index, row in data.iterrows():
    if row['mov'] == 'forward':
        forward_X = row['meters']
        d += forward_X * aim 
    if row['mov'] == 'up':
        aim = aim - row['meters']
    if row['mov'] == 'down':
        aim = aim + row['meters']

    print(row['mov'], row['meters'], aim, d)
```
